```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.plot.scatter(x='runtime',y='profit',c='red',s=50,figsize=(12,3))
```

```python
df.plot(x='Producer1', y='Producer2', style='o').astype(float)
plt.title('Producer1 vs Producer2')
plt.xlabel('Producer1')
plt.ylabel('Producer2')
plt.show()
df[['Producer1','Producer2']].corr()
```

```python
df.plot.scatter(x = 'Producer1',y = 'Producer2')
```
